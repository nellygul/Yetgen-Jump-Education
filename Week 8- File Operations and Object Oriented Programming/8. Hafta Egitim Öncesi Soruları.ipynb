````python
old_word = input("Enter the word to replace: ")
new_word = input("Enter the new word: ")
with open("file.txt", "r") as file:
    text = file.read()
new_text = text.replace(old_word, new_word)
with open("file.txt", "w") as file:
    file.write(new_text)
````
